```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read datasources and combine dataframes into one

```python
# Read the dataframes
aapl_df = pd.read_csv('./../AAPL_stock_data.csv')
amzn_df = pd.read_csv('./../AMZN_stock_data.csv')
googl_df = pd.read_csv('./../GOOGL_stock_data.csv')
msft_df = pd.read_csv('./../MSFT_stock_data.csv')
tsla_df = pd.read_csv('./../TSLA_stock_data.csv')
```

```python
# Combine all of the datasets
df = pd.concat([aapl_df, amzn_df, googl_df, msft_df, tsla_df])
df
```

```python
# Rename columns
df.columns = ['date', 'open', 'high', 'low', 'close', 'volume', 'symbol']
df
```

```python
# Save cleaned data
df.to_csv('./data.csv', index=False)
```

# Data validation and cleansing

```python
# Check for NULL values
df.isna().sum()
```

The dataset doesn't have any empty values

```python
# Plot boxplot to identify outliers
df.loc[:, 'open':'close'].boxplot();
```

```python
# Chart
grouped_df = df.groupby(['symbol', 'date'])['close'].agg('mean').reset_index()

symbols = grouped_df['symbol'].unique()
for symbol in symbols:
    aapl_df = grouped_df[grouped_df['symbol'] == symbol]
    x = aapl_df['date']
    x_points = np.arange(len(x))
    y = aapl_df['close']
    plt.plot(x_points, y)
plt.show()
```
